```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv("kidney_disease.csv")
Lesser=[]
Greater=[]
quan=[]
qual=[]
dataset
```

```python
#Assigning dataset to dataframe for Dropping the id column which is not necessary for analysis
df=pd.DataFrame(dataset)
dfid=pd.DataFrame(dataset['id'])
```

```python
df=df.drop(["id"],axis=1)
dataset=df
```

```python
#Splitting numerical and categorical data
from Univariatekidney import univar
u=univar(dataset)
qual,quan=u.quanqual()
```

```python
# In the dataset PCV,RC,WC are  umeric values but its typr is object so it is been converted to numeric value
dataset['pcv']=pd.to_numeric(dataset['pcv'],errors='coerce')
dataset['wc']=pd.to_numeric(dataset['wc'],errors='coerce')
dataset['rc']=pd.to_numeric(dataset['rc'],errors='coerce')
```

```python
#Finding the null values
dataset.isnull().sum()
```

```python
#Assigining median values to null values of numrical data
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy="median")
impute=impute.fit(dataset[quan])
imtable=pd.DataFrame(impute.transform(dataset[quan]).round(2))
dataset[quan]=imtable
```

```python
#Assigining median values to null values of categorical data
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
impute=impute.fit(dataset[qual])
imptable=pd.DataFrame(impute.transform(dataset[qual]))
dataset[qual]=imptable
```

```python
#Inserting the id column which has been dropped initially
df.insert(0,"id",dfid)
dataset=df
```

```python
# saving preprocessed data to new csv file
dataset.to_csv("prekidney.csv",index=False) 
```
